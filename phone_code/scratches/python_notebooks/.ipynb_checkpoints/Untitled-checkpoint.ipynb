```python
import pandas as pd
import altair as alt
```

```python
source = pd.DataFrame({
    'date': ['2/15', '2/16', '2/17', '2/18', '2/19', '2/20', 'Today'],
    'hours slept': [6.5, 7.1, 6.15, 6, 7.5, 8, 6]
})

chart = alt.Chart(source).mark_bar(size=20).encode(
    x='date',
    y='hours slept'
).properties(
    width=350,
    height=250
).configure_axis(
    labelFontSize=15,
    titleFontSize=15
)
chart.encoding.y.scale = alt.Scale(domain=[0, 10])
chart
```

```python
source = pd.DataFrame({
    'date': ['2/15', '2/16', '2/17', '2/18', '2/19', '2/20', 'Today'],
    'motivation': [1, 4, 4, 5, 3, 3, 5]
})

chart = alt.Chart(source).mark_line(point=True).encode(
    x='date',
    y='motivation'
).properties(
    width=350,
    height=250
).configure_axis(
    labelFontSize=15,
    titleFontSize=15
)
chart.encoding.y.scale = alt.Scale(domain=[-0.1, 5.2])
chart
```

```python
source = pd.DataFrame({
    'date': ['2/15', '2/16', '2/17', '2/18', '2/19', '2/20', 'Today'],
    'hours slept': [6.5, 7.1, 6.15, 6, 7.5, 8, 6],
    'motivation': [1, 4, 4, 5, 3, 3, 5]
})

chart = alt.Chart(source).mark_bar(size=20).encode(
    x='date',
    y='hours slept'
).properties(
    width=350,
    height=250
)
chart.encoding.y.scale = alt.Scale(domain=[0, 10])

chart2 = alt.Chart(source).mark_line().encode(
    x='date',
    y='motivation'
).properties(
    width=350,
    height=250
)
chart2.encoding.y.scale = alt.Scale(domain=[-0.1, 5.2])


(chart + chart2).configure_axis(
    labelFontSize=15,
    titleFontSize=15
) 
```
