```python
import pandas as pd

sleep_data = {
    'date': [1,2,3,4,5,6],
    'start_of_sleep': [2,3,1.5,3,1,0],
    'sleep_duration_in_hours': [6,7,5,6,8,9]
}
sleep_data['end_of_sleep'] = [sleep_data['start_of_sleep'][i]+sleep_data['sleep_duration_in_hours'][i] 
                              for i in range(len(sleep_data['date']))]


#-- convert to pandas dataframe, with time based indexing
sleep_data_frame = pd.DataFrame(sleep_data, columns = ['date','start_of_sleep','end_of_sleep', 'sleep_duration_in_hours']) 

sleep_data_frame
```

```python
import altair as alt
from vega_datasets import data

#this is for color good vs bad day sleep
open_close_color = alt.condition(alt.datum.sleep_duration_in_hours > 6,
                                 alt.value("#06982d"),
                                 alt.value("#ae1325"))

alt.Chart(sleep_data_frame).mark_bar().encode(
    x='date:O',
    y='start_of_sleep:Q',
    y2='end_of_sleep:Q',
    color=open_close_color
)
```
